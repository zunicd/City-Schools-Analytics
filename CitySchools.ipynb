```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Give the displayed data cleaner formatting

```python
# Total number of schools
tot_schools = len(school_data)
tot_schools
```

```python
# Total number of students
tot_students = len(student_data)
tot_students
```

```python
# Total budget
budget = school_data['budget'].sum()
budget
```

```python
# Average math score
avg_math = student_data['math_score'].mean()
avg_math
```

```python
# Average reading score
avg_read = student_data['reading_score'].mean()
avg_read
```

```python
# Overall passing rate
overall_pass = (avg_math + avg_read) / 2
overall_pass
```

```python
# Percentage of students with a passing math score (70 or greater)
pass_math = (student_data.loc[student_data['math_score'] >= 70]['Student ID'].count() / tot_students) * 100
pass_math
```

```python
# Percentage of students with a passing reading score (70 or greater)
pass_read = (student_data.loc[student_data['reading_score'] >= 70]['Student ID'].count() / tot_students) * 100
pass_read
```

```python
# Create dataframe
summary_df = pd.DataFrame({'Total Schools': [tot_schools],
                          'Total Students': tot_students,
                          'Total Budget': budget,
                          'Average Math Score': avg_math,
                          'Average Reading Score': avg_read,
                          '% Passing Math': pass_math,
                          '% Passing Reading': pass_read,
                          '% Overall Passing Rate': overall_pass})
summary_df.head()
```

```python
# Format to go to two decimal places, include a dollar sign, and use comma notation for revenue
summary_df['Total Students'] = summary_df['Total Students'].map('{:,d}'.format)
summary_df['Total Budget'] = summary_df['Total Budget'].map('${:,.2f}'.format)

summary_df.head()
```

```python
school_df = pd.DataFrame(school_s1)
school_df.head()
```

```python
# Keep only school_name as index
school_df = school_df.reset_index(level=['type', 'budget', 'size'])
school_df.head()
```

```python
# Move column "Total Students" to 2nd place (index 1)
school_df.insert(1, 'Total Students',school_df.pop('Total Students'))
school_df.head()
```

```python
# Calculate budget per student
budget_ps = school_df['Total School Budget'] / school_df['Total Students']

# Insert this series as 4th column (index 3) with given name
school_df.insert(3, 'Per Student Budget', budget_ps)

school_df.head()
```

```python
# school_gb = df.groupby(['school_name'])  # set earlier
school_gb['reading_score'].mean()
```

```python
# Calculate average reading score
school_df['Average Reading Score'] = school_gb['reading_score'].mean()
school_df.head()
```

```python
# Calculate % passing math series
pass_math_s = (school_data_df.loc[school_data_df['math_score'] >= 70, 'school_name'].value_counts() 
               / school_df['Total Students']) * 100
pass_math_s
```

```python
# Add above series to dataframe
school_df['% Passing Math'] = pass_math_s
school_df.head()
```

```python
# Calculate % passing reading series
pass_read_s = (school_data_df.loc[school_data_df['reading_score'] >= 70, 'school_name'].value_counts() 
               / school_df['Total Students']) * 100
pass_read_s
```

```python
# Add above series to dataframe
school_df['% Passing Reading'] = pass_read_s
school_df.head()
```

```python
# Calculate % overall passing rate and add it as column
school_df['% Overall Passing Rate'] = (school_df['% Passing Math'] + 
                                       school_df['% Passing Reading']) / 2
school_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_df_spending = school_df.copy()
school_df_spending.head()
```

```python
# Spending bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Slice data and place it into bins
pd.cut(school_df_spending['Per Student Budget'], spending_bins, labels=group_names)
```

```python
# Append spending bins to spending dataframe
school_df_spending['Spending Ranges (Per Student)'] = \
        pd.cut(school_df_spending['Per Student Budget'], spending_bins, labels=group_names)

school_df_spending.head()
```

```python
# Group by spending ranges per stugent
school_df_spending = school_df_spending.groupby('Spending Ranges (Per Student)')\
            [['Average Math Score', 'Average Reading Score', '% Passing Math', 
              '% Passing Reading', '% Overall Passing Rate']].mean()

school_df_spending
```

## Observable Trends

There are several trends that we could observe based on the analyzed data

* **Money does not matter** – It looks that school performances are better for schools with lower spending per student. There is especially a huge difference with passing rates, around 95% for schools with lower spending comparing to around 80% and even around 70% for schools with higher spending.


* **Size does matter** – School performances are much better for small and medium size schools than for large size schools.


* **Type does matter** – School performances for Charter schools are much better than for District schools.
